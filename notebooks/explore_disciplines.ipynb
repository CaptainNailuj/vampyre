```python
import os
import json
from tqdm import tqdm
import itertools
```

```python
path_disciplines = "../util_files/disciplines.json"
dict_disciplines = json.load(open(path_disciplines))
```

```python
dict_disciplines_corrected = {}
for discipline in tqdm(dict_disciplines):
    dict_levels = {}
    for level in dict_disciplines[discipline]:
        if '-' in level:
            start_level, end_level = [int(x) for x in level.split('-')]
            dict_levels.update({i: [dict_disciplines[discipline][level]] for i in range(start_level, end_level+1)})
            continue

        level_int = int(level.split('_')[0])
        powers_list = dict_levels.get(level_int, [])
        powers_list.append(dict_disciplines[discipline][level].lower())
        dict_levels[level_int] = powers_list
        
    dict_disciplines_corrected[discipline.lower()] = dict_levels
    # break

dict_disciplines_corrected
```
